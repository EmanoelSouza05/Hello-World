```python
class Carro():

    def __init__(self, Marca, Modelo, Ano):

        self.Marca = Marca
        self.Modelo = Modelo
        self.Ano = Ano
        self.Ligado = False


    def Ligar(self):
        self.Ligado = True
        if self.Ligado == True:
            print("Carro já está ligado")
        else:
            print("Carro ligado")


    def get_velocidade(self):
        velocidade = 0
        print(f"\nA velocidade atual do carro é = {velocidade}")


    def set_Aceleração(self):
        valor = int(input())
        velocidade += valor
        return(f"\nO carro acelerou {valor}m/s")

    def desligar(self):
        self.Ligado = False
        if self.Ligado == False:
            return("Carro já está desligado")
        else:
            return("Carro desligado")


    def info(self):
        print(Marca.capitalize(), Modelo.capitalize(), Ano.capitalize()) 
        
```

```python
car = Carro("Toyota", "Supra", 1985)
```

```python
car.set_Aceleração()
```
